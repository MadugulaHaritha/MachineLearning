```python
data=data.drop(['Column1','Column2','Column3','Column4','status_published'],axis=1)
data
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoded_cols = encoder.fit_transform(data[['status_type']])

# Get feature names from the encoder categories
feature_names = encoder.categories_[0]

# Create DataFrame with one-hot encoded columns
encoded_df = pd.DataFrame(encoded_cols, columns=feature_names)

# Concatenate with original DataFrame
final_df = pd.concat([data, encoded_df], axis=1)

print(final_df)
```

```python
x=data.drop('status_type',axis=1)
y=data['status_type']
print(x.head())
print(y.head())
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
y
```

```python
#import train_test_split function
from sklearn.model_selection import train_test_split
# Split dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
#import adaboost classifier
from sklearn.ensemble import AdaBoostClassifier

#create adaboost classifier object
abc=AdaBoostClassifier(n_estimators=50, learning_rate=1,random_state=0)


#train adaboost classifier
model=abc.fit(x_train,y_train)
data.head()
```

```python

#predict the response for test dataset
y_pred=model.predict(x_test)
y_pred
```

```python
#import scikit.learn metrics module for accuracy calculation
from sklearn.metrics import accuracy_score

#calculate and print model accuracy
print("Xgboost classifier model Accuracy:",accuracy_score(y_test, y_pred))
```
